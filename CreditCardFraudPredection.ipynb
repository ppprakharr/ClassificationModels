Under Sampling

Build a smaple dataset containing similar distribution of normal transaction and fraudulant transaction

Number of fraud transaction = 482

```python
legit_sample = legit.sample(n=482)
```

Concatinating 2 dataframes
legit_sample and fraud

```python
new_dataset = pd.concat([legit_sample, fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
# check the distribution in the new dataset
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the dataset into label and target

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
print(x)
print(y)
```

Splitting training ad test data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1,random_state=8, stratify=y)
```

```python
# training the data
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
# evaluating the training data
train_data_predict = model.predict(x_train)
accuracy_check = accuracy_score(train_data_predict, y_train)
print("Accuracy score for train: ",accuracy_check)
```

```python
# evaluating the test data
test_data_predict = model.predict(x_test)
accuracy_check = accuracy_score(test_data_predict, y_test)
print("Accuracy score for test: ",accuracy_check)
```
