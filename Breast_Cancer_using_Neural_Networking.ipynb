#IMPORTING THE DEPENDENCIES

```python
import numpy as np
import pandas as pd
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

Data collectetion and preprocessing

```python
cancer_data = sklearn.datasets.load_breast_cancer()
```

```python
print(cancer_data)
```

```python
data_frame = pd.DataFrame(cancer_data.data, columns = cancer_data.feature_names)
```

```python
# first 5 eleements
data_frame.head()
```

```python
data_frame['label']=cancer_data.target
```

```python
# printing tail
data_frame.tail()
```

```python
data_frame.shape
```

```python
# check for missing value
data_frame.isnull().sum()
```

```python
# statistival measure
data_frame.describe()
```

1-->non cancer

0---> cancer

```python
data_frame.groupby('label').mean()
```

splitting the feature and label

```python
x=data_frame.drop(columns = 'label',axis=1)
y=data_frame['label']
print(x)
print(y)
```

Splitting the data into training and testing data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

Standardise the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train_std = scaler.fit_transform(x_train)
x_test_std= scaler.transform(x_test)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Building the Neural Network

```python
# importing tesorflow and keras
import tensorflow as tf
```

```python
tf.random.set_seed(3)
```

```python
from tensorflow import keras
```

```python
#setting up layers of the neural network
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),     #contains the number of columms
                          keras.layers.Dense(20,activation='relu'),     #relu=rectified linear unit
                          keras.layers.Dense(2, activation='sigmoid')

])
```

```python
# compiling the neural network
model.compile(optimizer='adam',
loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the neural network
history=model.fit(x_train_std,y_train,validation_split=0.1,epochs=10)
```

Visualising the accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epochs')
```

To Get the accuracy value on test data

```python
loss, accuracy = model.evaluate(x_test_std,y_test)
print('accuracy value: ',accuracy)
```

Building the predicitve system

```python
input_data=(x_test_std[2])
input_data_numpy = np.asarray(input_data)
inp_reshape = input_data_numpy.reshape(1,-1)
#standardising
inpu_reshape_std = scaler.transform(inp_reshape)
prediction=model.predict(inpu_reshape_std)

print(prediction)

#applying argmax
prediction_label=[np.argmax(prediction)]

if (prediction_label[0]==0):
  print('The tumour is cancerous')
else:
  print('the tumor is non-cancerous')
```
